# Data size finding

This notebook will try to determine if higher quality image means better results 

**Main goals are**: 
- Find a good resolution
- Conclude on image resolution's impact on model performances

```python
# Retreiving a significative amount of data
from local_lib.dataset import retreiveDatasetFromCsv

dataset = retreiveDatasetFromCsv(src="../../datasets/chest_Xray/_processed_resize_small/_processed_dfs")
dataset = dataset.sample(frac=1, random_state=42)
```

**We apply the same treatment for all tries**

```python
from local_lib.flatten_transformer import FlattenTransformer
from sklearn.pipeline import Pipeline
import numpy as np

pipeline = Pipeline([
    ('image_mask', FlattenTransformer()),
])

# Fit and transform data
X = pipeline.fit_transform(dataset["pixel_value"])
y = np.array(list(dataset["class"]))
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

**Global results**

```python
train_accuracys = []
train_precisions = []
train_recalls = []
test_accuracys = []
test_precisions = []
test_recalls = []
```

**220x163px**

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import recall_score, precision_score, accuracy_score

pipeline = Pipeline([
    ('knn', KNeighborsClassifier())
])

pipeline.fit(X_train, y_train)

y_train_pred = pipeline.predict(X_train)
y_test_pred = pipeline.predict(X_test)

# Compute metrics for the training set
train_accuracy = accuracy_score(y_train, y_train_pred)
train_precision = precision_score(y_train, y_train_pred, average="weighted")
train_recall = recall_score(y_train, y_train_pred, average="weighted")

# Compute metrics for the test set
test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred, average="weighted")
test_recall = recall_score(y_test, y_test_pred, average="weighted")

train_accuracys.append(train_accuracy)
train_precisions.append(train_precision)
train_recalls.append(train_recall)
test_accuracys.append(test_accuracy)
test_precisions.append(test_precision)
test_recalls.append(test_recall)

print(f"Training Accuracy: {train_accuracy:.2f}")
print(f"Training Precision: {train_precision:.2f}")
print(f"Training Recall: {train_recall:.2f}")

print(f"--------")

print(f"Test Accuracy: {test_accuracy:.2f}")
print(f"Test Precision: {test_precision:.2f}")
print(f"Test Recall: {test_recall:.2f}")

print(f"--------")

disp = ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred)
disp.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix on test data:\n")
plt.show()
```
